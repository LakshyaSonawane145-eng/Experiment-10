```python
#Introduction to Pandas
#Pandas : Panel Data
import pandas as pd

s = pd.Series([10,20,30,40])
print(s)
```

```python
#Create DataFrame
data = {
    "Name":["A","B","C"],
    "Marks":[81,90,78]
}

df = pd.DataFrame(data)
print(df)
```

```python
#Find the Structure of the DataFrame
print(df.shape)
print(df.ndim)
print(df.size)
```

```python
#Display Column Names
print(df.columns)
```

```python
#Check Data Types of Each Column
print(df.dtypes)
```

```python
#Access Individual Column
print(df["Name"])
print(df["Marks"])
df.loc[0,"Name"]
```

```python
df = pd.DataFrame(data)
print(df)
print(df.loc[0,"Name"])
```

```python
#Add a New Column

df["Grade"] = ["First Class" , "Distinction" , "Second Class"]
print(df)
```

```python
#Update Data in DataFrame
#Change marks of student A to 88

df.iloc[0,0] = 88
print(df)
```

```python
#Delete a Column
df1 = df.drop("Grade",axis=1)
print(df)
print(df1)
```

```python
#Perform Basic Statistical Analysis
mean1 = df["Marks"].mean()
print(mean1)
```

```python
min1 = df["Marks"].min()
print(min1)
```
